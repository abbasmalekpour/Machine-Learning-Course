````python
inv(A.T @ A) @ A.T  # A+ (pseudo-inverse)
````

If $A$ is non-singular:

$$A^{+} = (A^T A)^{-1} A^T = A^{-1} (A^T)^{-1} A^T = A^{-1} [(A^T)^{-1} A^T] = A^{-1} I = A^{-1}$$

````python
inv(A)
````

````python
pinv(A)
````

````python
np.allclose(inv(A), pinv(A))
````

## Example: Linear Regression Cost Function

**Goal:** Given $X, y, \theta$, compute $J(\theta)$.

$$J(\theta) = \frac{1}{2} \sum_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)}) ^ 2 = \frac{1}{2} \sum_{i=1}^{m} (\theta^T x^{(i)} - y^{(i)}) ^ 2 = \frac{1}{2} (X \theta - y)^T (X \theta - y)$$

```python
    def sse(X, y, theta):
        """ Sum of Squared Error cost function.
        """
        ...
        return cost
```

````python
X = np.array([[1, 1], [1, 2], [1, 3]])
y = np.array([[1], [2], [3]])
theta = np.array([[1], [0]])
````

````python
plt.style.use('ggplot')

def plot_hyp(X, y, theta):
    # plot data
    plt.scatter(X[:, 1], y, s=100, c='r', marker='x')
    
    x = np.linspace(0.8, 3.2)
    x = x[:, None]
    x = np.hstack((np.ones_like(x), x))
    
    # plot hypothesis
    h = x @ theta
    plt.plot(x[:, 1], h, c='k', lw=2)
    plt.show()
````

````python
plot_hyp(X, y, theta)
````

````python
def sse(X, y, theta):
    """A vectorized implementation of Sum of Squared Error.
    """
    predictions = X @ theta
    errors = predictions - y
    return 0.5 * np.sum(errors ** 2)
````

````python
def sse(X, y, theta):
    """A vectorized implementation of Sum of Squared Error.
    """
    errors = X @ theta - y
    return 0.5 * (errors.T @ errors)[0, 0]
````

````python
theta = np.array([[0], [0]])

plot_hyp(X, y, theta)
loss = sse(X, y, theta)
print("J(%s) = %.2f" %(theta.ravel(), loss))
````

````python
theta = np.array([[0], [0.5]])

plot_hyp(X, y, theta)
loss = sse(X, y, theta)
print("J(%s) = %.2f" %(theta.ravel(), loss))
````

````python
theta = np.array([[0], [1.0]])

plot_hyp(X, y, theta)
loss = sse(X, y, theta)
print("J(%s) = %.2f" %(theta.ravel(), loss))
````

````python
theta = np.array([[0], [1.5]])

plot_hyp(X, y, theta)
loss = sse(X, y, theta)
print("J(%s) = %.2f" %(theta.ravel(), loss))
````

````python
theta = np.array([[0], [2.0]])

plot_hyp(X, y, theta)
loss = sse(X, y, theta)
print("J(%s) = %.2f" %(theta.ravel(), loss))
````

````python
theta_1s = np.linspace(-1, 3, num=50)
J = []
````

````python
for theta_1 in theta_1s:
    cost = sse(X, y, np.array([[0], [theta_1]]))
    J += [cost]
````

````python
plt.plot(theta_1s, J)

plt.plot(1, 0, 'ro')
plt.xlabel(r'$\theta$')
plt.ylabel(r'$J(\theta)$')

plt.annotate('Global Optimum',
             xy=(1, 1), xycoords='data',
             xytext=(-15, 50), textcoords='offset points',
             arrowprops=dict(facecolor='black', shrink=0.05),
             horizontalalignment='center', verticalalignment='bottom')

plt.grid(True)
plt.show()
````

````python
with plt.xkcd():
    plt.plot(theta_1s, J)
    plt.plot(1, 0, 'ro')
    plt.xlabel(r'$\theta$')
    plt.ylabel(r'$J(\theta)$')

    plt.annotate('Global Optimum',
                 xy=(1, 1), xycoords='data',
                 xytext=(-15, 50), textcoords='offset points',
                 arrowprops=dict(facecolor='black', shrink=0.05),
                 horizontalalignment='center', verticalalignment='bottom')
    plt.show()
````
